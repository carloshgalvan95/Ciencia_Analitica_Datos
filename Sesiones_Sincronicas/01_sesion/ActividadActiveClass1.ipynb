```python
# Con groupby() se forman particiones o grupos, de manera que dentro de cada grupo todas las filas tengan el mismo valor en la columna especificada
# En esta caso, saca el promedio por industria para todas las columnas numéricas del dataframe
forbes_df.groupby('Industry').mean(numeric_only=True)
```

```python
# Si estamos interesados en una columna en particular, se puede especificar tras groupby() para que las operaciones (como mean()) se apliquen únicamente a esa columna
forbes_df.groupby('Industry')['Assets (Billions)'].mean()
```

```python
# El dataframe original está en formato ancho como se muestra a continuación
forbes_df.head()
```

```python
# pd.melt() transforma un DataFrame de formato ancho a largo
# - id_vars: columnas que se mantienen como identificadores (no se “derriten”)
# - var_name: nombre de la nueva columna que contendrá los nombres de las columnas originales “derretidas”
# - value_name: nombre de la nueva columna que contendrá los valores correspondientes
forbes_df.reset_index(inplace=True)
forbes_long = forbes_df.melt(id_vars=['2022 Ranking','Organization Name', 'Industry', 'Country', 'Year Founded', 'CEO'], var_name='Metric', value_name='Value')
forbes_long
```

```python
# pd.pivot() transforma un DataFrame de formato largo a ancho
# - index: columnas que identifican cada fila en el DataFrame ancho resultante
# - columns: columna cuyos valores únicos se convierten en nuevas columnas
# - values: columna que contiene los valores que se colocarán en las celdas del DataFrame ancho
forbes_wide = forbes_long.pivot(index=['2022 Ranking','Organization Name', 'Industry', 'Country', 'Year Founded', 'CEO'], columns='Metric', values='Value').reset_index()
forbes_wide.set_index('2022 Ranking', inplace=True)
forbes_wide
```

```python
# pd.stack() convierte columnas en filas creando un MultiIndex en las filas, transformando el dataFrame ancho en largo
forbes_multi = forbes_wide.stack()
forbes_multi.head(20)
```

```python
# pd.unstack(): Desapila un nivel del índice de filas hacia columnas, transformando el dataFrame largo en ancho
forbes_multi.unstack()
```

```python
# En el método plot() se especifica el tipo de gráfico a construir con el parámetro kind
# Cada columna representa una clase o bin (incluye un rango de valores)
forbes_df['Revenue (Billions)'].plot(kind='hist')
plt.show()
```

```python
# Lo anterior es equivalente a colocar el tipo de gráfico después del plot y separado por un punto
forbes_df['Revenue (Billions)'].plot.hist()
plt.show()
```

```python
# Por defecto se dibujan 10 bins, aunque en el gráfico anterior no se alcanzan a apreciar por el sesgo
# Pero este número se puede cambiar
forbes_df['Revenue (Billions)'].plot(kind='hist', bins=20)
plt.show()
```

```python
# También se pueden enumerar los límites de todos los bins, sin necesidad de que sean equidistantes
forbes_df['Revenue (Billions)'].plot(kind='hist', bins=[0,25,50,100,200])
plt.show()
```

```python
# Si son equidistantes, se pueden indicar con el método arange(inicio, fin, incremento) de numpy
forbes_df['Revenue (Billions)'].plot.hist(bins=np.arange(0,200,25))
plt.show()
```

```python
# El método hist() del dataframe, obtiene un histograma para cada columna numérica
# Se puede ajustar el tamaño con figsize
forbes_df.hist(figsize=(12,10))
plt.show()
```
